```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/creditcard/creditcard.csv')

dataset.columns
```

```python
# Viewing the dataset we have for better understanding

dataset['Class'].value_counts()
```

```python
dataset[['Time','Amount']].hist(bins = 30)#, figsize=(10,10))
```

```python
data = dataset.copy()

# 1. Descriptive Statistics
desc_stats = data.describe()

# 2. Class Distribution
class_distribution = data['Class'].value_counts()

# 3. Correlation Analysis
correlation_matrix = data.corr()

# Plotting
# 4. Distribution of 'Amount' and 'Time'
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(data['Amount'], kde=True, bins=50)
plt.title('Distribution of Transaction Amount')

plt.subplot(1, 2, 2)
sns.histplot(data['Time'], kde=True, bins=50, color='orange')
plt.title('Distribution of Transaction Time')

plt.tight_layout()
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()

desc_stats, class_distribution
```

```python
# Outlier Detection using Boxplots
plt.figure(figsize=(14, 6))

# Boxplot for 'Amount'
plt.subplot(1, 2, 1)
sns.boxplot(x=data['Amount'])
plt.title('Boxplot of Transaction Amount')

# Boxplot for 'Time'
plt.subplot(1, 2, 2)
sns.boxplot(x=data['Time'], color='orange')
plt.title('Boxplot of Transaction Time')

plt.tight_layout()
plt.show()
```

```python
# Visualize Class Imbalance
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=data, palette=['green', 'red'])
plt.title('Class Distribution (Legitimate vs. Fraudulent Transactions)')
plt.xticks([0, 1], ['Legitimate (0)', 'Fraudulent (1)'])
plt.ylabel('Count')
plt.show()
```
